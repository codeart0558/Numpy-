```python
# normal distribution:-It is also called the Gaussian Distribution after the German mathematician Carl Friedrich Gauss.
# Use the random.normal() method to get a Normal Data Distribution.
# It fits the probability distribution of many events, eg. IQ Scores, Heartbeat etc.
```

```python
# It has three parameters:

# loc - (Mean) where the peak of the bell exists.

# scale - (Standard Deviation) how flat the graph distribution should be.

# size - The shape of the returned array.
```

```python
# 1. loc

# Meaning: The mean (center) of the distribution — the “peak” of the bell curve.

# Effect: Moves the curve left or right along the X-axis.

# Example:

# loc=0 → Bell centered at 0

# loc=50 → Bell centered at 50
# -------------------------------------------------------------------
# 2. scale

# Meaning: The standard deviation — how spread out the data is.

# Effect:

# Small scale (e.g., scale=1) → narrow, tall peak.

# Large scale (e.g., scale=10) → wide, flat curve.
# --------------------------------------------------------------------
# 3. size

# Meaning: The number of values (or shape) to return.

# Can be:

# Single number → returns a 1D array

# Tuple → returns multi-dimensional array

# Example:

# size=10 → 1D array with 10 values

# size=(2,3) → 2 rows × 3 columns array
```

```python
import numpy as np
from numpy import random
# same as
# arr = random.rand(2,3)
# print(arr)

arr = random.normal(size=(2,3))
print(arr)
```

```python
# Generate a random normal distribution of size 2x3 with mean at 1 and standard deviation of 2:
arr = random.normal(loc=1,scale=2,size=(2,3))
print(arr)
```

```python
# Visualization of Normal Distribution
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# for histograme use this
sns.displot(random.normal(size=1000))
plt.show()

# for curve kde use this
sns.displot(random.normal(size=1000),kind="kde")
plt.show()
```

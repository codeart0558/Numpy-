```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot([0,1,2,3,4,5])
plt.show()
```

```python
# A histogram is a type of plot that shows the distribution of
# numerical data â€” in other words, it tells you how often values fall 
# into different ranges (called bins).

# X-axis â†’ Represents ranges (intervals) of data values.

# Y-axis â†’ Represents the frequency (how many values fall into each range).

# Itâ€™s like a bar chart, but instead of categories, it uses value ranges.

# Helps you understand the shape of data (normal, skewed, etc.).


# Displot - distribution plot(curv plat - looks like histogram)
```

```python
sns.displot([0, 1, 2, 3, 4, 5], kind="kde")
plt.show()
```

```python

# Histogram
sns.histplot([0,1,2,3,4,5])
plt.show()

# KDE plot
sns.kdeplot([0,1,2,3,4,5])
plt.show()
```

```python


data = [0, 1, 2, 3, 4, 5]

sns.histplot(data, kde=True)  # Show histogram + KDE curve
plt.show()
```
